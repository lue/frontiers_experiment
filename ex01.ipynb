```python
i = 10
d1 = fits.open(files[i])
i = 14
d2 = fits.open(files[i])
plt.figure(figsize=(12,12))
plt.imshow((d1[1].data[2000:3000,1700:2700] - d2[1].data[2000:3000,1700:2700]), origin='lower')
```

```python
import scipy.misc
scipy.misc.imsave('test1.tiff', d1[1].data[2000:3000,1700:2700])
scipy.misc.imsave('test2.tiff', d2[1].data[2000:3000,1700:2700])
# >>> ar2 = scipy.misc.imread('scipy.tiff')
# >>> ar2.dtype


```

```python
import SimpleITK as sitk

elastixImageFilter = sitk.ElastixImageFilter()
elastixImageFilter.SetFixedImage(sitk.ReadImage("test1.tiff"))
elastixImageFilter.SetMovingImage(sitk.ReadImage("test2.tiff"))
```

```python
# dir(elastixImageFilter.SetFixedImage)
```

```python

parameterMapVector = sitk.VectorOfParameterMap()
parameterMapVector.append(sitk.GetDefaultParameterMap("affine"))
# parameterMapVector.append(sitk.GetDefaultParameterMap("bspline"))
elastixImageFilter.SetParameterMap(parameterMapVector)

elastixImageFilter.Execute()
```

```python
sitk.WriteImage(elastixImageFilter.GetResultImage(), "test.tiff")
```
